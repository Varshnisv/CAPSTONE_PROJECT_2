## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
# Step 1 Handle Missing Values
App_DATA = App_DATA[~App_DATA['Installs'].str.contains('Free')]
App_DATA.dropna(subset=['Rating', 'Size'], inplace=True)
App_DATA['Rating'].fillna(App_DATA['Rating'].mean(), inplace=True)

#Step 2 Change or convert all datatypes in App dataset
App_DATA['Installs'] = App_DATA['Installs'].str.replace('[+,]', '', regex=True).astype(int)
App_DATA['Price'] = App_DATA['Price'].str.replace('$', '').astype(float)
App_DATA['Size'] = App_DATA['Size'].replace('Varies with device', np.nan)
App_DATA['Size'] = App_DATA['Size'].str.replace('M', '').str.replace('k', '').astype(float)
App_DATA['Reviews'] = App_DATA['Reviews'].astype(int)
```

```python
#Step 3
# Feature Engineering
App_DATA['Is_Free'] = App_DATA['Type'].apply(lambda x: 1 if x == 'Free' else 0)

# Remove Duplicates
App_DATA.drop_duplicates(inplace=True)

# Clean Customer Reviews Data
Rev_DATA.drop_duplicates(inplace=True)
```

```python
# Merging Apps and Reviews Data
merged_data = pd.merge(App_DATA, Rev_DATA, on='App', how='inner')

# Handle missing values in the merged data
merged_data.fillna({'Sentiment': 'Neutral', 'Sentiment_Polarity': 0, 'Sentiment_Subjectivity': 0}, inplace=True)

# Dataset is now ready for analysis!
print("\nApps Dataset after cleaning and transformation:")
print(App_DATA.info())

print("\nCustomer Reviews Dataset after cleaning:")
print(Rev_DATA.info())

print("\nMerged Dataset (if required for analysis):")
print(merged_data.info())
```

#### Chart - 14 - Correlation Heatmap

```python
# Correlation Heatmap visualization code
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


features_extracted = ['Rating', 'Size', 'Installs', 'Price', 'Reviews']

correlation = App_DATA[features_extracted].corr()


plt.figure(figsize=(15, 9))

sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)


plt.title('Correlation Heatmap')
plt.show()
```

##### 1. Why did you pick the specific chart?

A correlation heatmap is useful for understanding the correlations between numerical characteristics in a dataset. This heatmap uses colour intensities to graphically illustrate the correlation matrix of important metrics (such as Rating, Size, Installs, Price, and Reviews), allowing you to quickly spot significant positive or negative relationships. The colour gradient highlights strong and low correlations, providing fast understanding into patterns.


##### 2. What is/are the insight(s) found from the chart?

The correlation heatmap shows the degree and direction of the associations between the features. For example:

A strong positive connection between installations and Reviews indicates that programs with more installations have more user reviews, which is expected but useful to validate.
Low or negative correlations, such as those between Price and Installs, may suggest that more expensive apps receive fewer downloads, implying a preference for free or inexpensive programs.
These information can help inform strategic decisions about app price, feature focus, and user acquisition techniques.


#### Chart - 15 - Pair Plot

```python
# Pair Plot visualization code
import seaborn as sns
import matplotlib.pyplot as plt


features = ['Rating', 'Size', 'Installs', 'Price', 'Reviews']


sns.pairplot(App_DATA[features])

plt.suptitle('Pair Plot', y=1.02)
plt.show()
```

A pair plot is used to visualise relationships between numerous pairs of numerical information at the same time. This map displays probable linear or non-linear connections, distributions, and correlations between parameters such as Rating, Size, Installs, Price, and Reviews. It provides a full picture of all pairwise interactions, which might uncover hidden patterns or clusters within the data.


##### 2. What is/are the insight(s) found from the chart?

The pair plot displays diagonal distributions as well as scatter plots for each paired feature combination. These insights might include:

Patterns or clusters between Installs and Reviews, indicating a positive link in which more installs typically result in more reviews.
The distributions of attributes like as Rating and Price illustrate how values are distributed, which might suggest user preferences or app-market trends.
Potential outliers in attributes like Price may suggest uncommon, high-priced apps that break from the overall trend.
These insights assist in understanding how characteristics interact and behave together, showing patterns and finding outliers.
